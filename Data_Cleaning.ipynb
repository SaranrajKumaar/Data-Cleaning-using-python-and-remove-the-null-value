# **Taking sample data sets in kaggle website**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
```

```python
file  = '/content/SAMPLE.csv'
```

```python
sample_pd = pd.read_csv(file)
sample_pd
```

# **How many null value are proesent**

```python
sample_pd.isnull().sum()
```

# **Detecting the Nan values**

```python
missing_values =['NaN',np.nan]

sample_pd = pd.read_csv(file,na_values=missing_values)
sample_pd
```

```python
sample_pd.isna().sum()
```

# **Use heatmap how many null are present using heatmap**

```python
sns.heatmap(sample_pd.isna(),yticklabels=False)
```

```python
sns.heatmap(sample_pd.isna(),yticklabels=False,annot=True)
```

# **How to remove Nan values**

```python
sample_pd.dropna(how='all')
```

```python
df_dropped = sample_pd.dropna(how='all',inplace=True)
df_dropped
```

```python
df_dropped.isna().sum()
```

```python

sample_pd.fillna(method='bfill')
sample_pd
```

```python
sample_pd.dropna(how='any',inplace=True)
```

```python
sns.heatmap(sample_pd.isna(),yticklabels=True)
```
